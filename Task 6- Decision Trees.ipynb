```python
#Splitting of the data into training and testing
#First we separate the data into X and Y variables
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
Y = data['Species'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, train_size=0.7)
```

```python
#Building the classifier
from sklearn.tree import DecisionTreeClassifier
dfc = DecisionTreeClassifier()
dfc.fit(X_train, Y_train)
```

```python
#Checking the accuracy of the model with the help of testing data
score = dfc.score(X_test, Y_test)
print(score)
```

The accuracy score of our model is pretty high

```python
import sklearn.datasets as datasets
iris=datasets.load_iris()
# Visualize the graph
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dfc, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```
